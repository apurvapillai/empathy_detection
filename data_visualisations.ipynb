```python

import pandas as pd
import numpy as np

train_file = '/content/updated_empathetic_dialogues_train.csv'
val_file = '/content/updated_empathetic_dialogues_val.csv'
test_file = '/content/updated_empathetic_dialogues_test.csv'

train_df = pd.read_csv(train_file, on_bad_lines='skip')
val_df = pd.read_csv(val_file, on_bad_lines='skip')
test_df = pd.read_csv(test_file, on_bad_lines='skip')


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

context_counts = train_df['context'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=context_counts.index, y=context_counts.values, palette='viridis')
plt.title('Conversations per Emotional Category (training)')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()


context_counts = test_df['context'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=context_counts.index, y=context_counts.values, palette='viridis')
plt.title('Conversations per Emotional Category (test)')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()

```

```python

from wordcloud import WordCloud

text = ' '.join(train_df['context'].dropna().unique())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Context')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
train_df['context'].value_counts().plot(kind='line', marker='o', color='purple')
plt.title('Trends of Emotional Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```
